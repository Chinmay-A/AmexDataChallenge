```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv('training_set.csv')
df.head()
```

```python
df.columns
```

```python
df['winner']=(df['winner_id']==df['team1_id'])
```

```python
winnersarr=[]

for i in range(len(df)):

    if(df['winner'][i]):
        winnersarr.append(1)
    else:
        winnersarr.append(0)

df['winners']=winnersarr
```

```python
df.head()
```

```python
toss_code=[]

for i in range(len(df)):
    if(df['toss winner'][i]!=df['team1'][i] and df['toss winner'][i]!=df['team2'][i]):
        print("Issue Detected in row: ")
        print(i)
    if(df['toss winner'][i]==df['team1'][i] and df['toss decision'][i]=="bat"):
       print("OK")
       toss_code.append(1)
    elif(df['toss winner'][i]==df['team2'][i] and df['toss decision'][i]=="field"):
        print("OKA")
        toss_code.append(1)
    else:
        print("BHA")
        toss_code.append(0)

df['tossA']=toss_code
```

```python
relevant_columns=['match id','team1_id','team2_id','team_count_50runs_last15','team_winp_last5','team1only_avg_runs_last15','team1_winp_team2_last15','ground_avg_runs_last15','momentum_score','location_score','batting_score','bowling_score','winners']
```

```python
df_new=df[relevant_columns]
df_new.to_csv('train_ready.csv',index=False)
```
